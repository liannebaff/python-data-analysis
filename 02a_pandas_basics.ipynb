### Locating DF Data

#### loc

```python
# df.loc[rows, column name(S)]
# df.iloc[rows, column idex(es)]

#df.loc uses column names and iloc uses column indexes
```

```python
# To retrun all rows but only the name column
df.loc[0:34,"name"]
```

```python
# To retrun 5 rows but only the name column
df.loc[0:4,"name"]
```

```python
# To return 5 rows and the name and class columns
df.loc[0:4,["name", "class"]]
```

```python
# To return all rows and the first column, don't write start and end ranges just use ":"
df.loc[:,"name"]
```

```python
# To return up to index 10
# If you don't mention starting point it will start from index 0
df.loc[:10,"name"]
```

```python
# To return all rows starting from index 10
# If you don't mention ending point it will end at the final index
df.loc[10:,"name"]
```

#### iloc

```python
# id is index 0
# name is index 1
# class is index 2
```

```python
# To return all the rows and the name column
df.iloc[:,1]
```

```python
# To return multiple columns they must be in box brackets
df.iloc[:,[1,2]]
```

```python
# To return first 5 rows
# In iloc stopping range is excluded in the output, therefore 0:5 would return until index 5 and doesn't include index 5.
df.iloc[0:5, [1,2]]
```

```python
# To retrun 5 rows and 2 columns, you can use ranges
df.iloc[0:5, 0:2]

```

### How to Filter Data


```python
#This is called boolean filtering.
# It will return true and false

df["mark"]>90
```

```python
# To ouput students with matks over 90.
df[df["mark"]>90]
```

```python
# To ouput children from class 6
df[df["class"]=="Six"]
```

```python
# Filter by multiple conditions
df[(df["class"]=="Six")& (df["gender"]=="male")]
```

```python
df[(df["gender"]=="female") & (df["mark"]>=85)]
```

```python
#& "and" operator
# | is called a pipe and is used for "or" operator

df[(df["gender"]=="female") | (df["mark"]>=90)]
```

```python
#df.loc[condition,column labels]
#iloc cannot be used to filter
df.loc[df["id"]<5, : ]
```
